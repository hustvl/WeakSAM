```python
import os

GT_VIS_PATH = "/data/zhulianghui/yingyueli/Mask2Former/cim-visualization/gt-visualization"
CIM_VIS_PATH = "/data/zhulianghui/yingyueli/Mask2Former/cim-visualization/cim-visualization"
WKS_VIS_PATH = "/data/zhulianghui/yingyueli/Mask2Former/output/voc12_insseg_maskformer2_R50_bs16_50ep/panoptic_results"

# 分别遍历三个 Path 下的所有图片
gt_vis_list = os.listdir(GT_VIS_PATH)
cim_vis_list = os.listdir(CIM_VIS_PATH)
wks_vis_list = os.listdir(WKS_VIS_PATH)

print("length of gt_vis_list: ", len(gt_vis_list))
print("length of cim_vis_list: ", len(cim_vis_list))
print("length of wks_vis_list: ", len(wks_vis_list))
```

```python
import cv2
from PIL import Image
import numpy as np
from google.colab.patches import cv2_imshow

# 分别可视化三个 Path 下的所有图片
for i in range(len(gt_vis_list)):
    gt_vis_img = cv2.imread(os.path.join(GT_VIS_PATH, gt_vis_list[i]))
    cim_vis_img = cv2.imread(os.path.join(CIM_VIS_PATH, cim_vis_list[i]))
    wks_vis_img = cv2.imread(os.path.join(WKS_VIS_PATH, wks_vis_list[i]))
    print("gt vis image path: ", os.path.join(GT_VIS_PATH, gt_vis_list[i]))
    print("cim vis image path: ", os.path.join(CIM_VIS_PATH, cim_vis_list[i]))
    print("wks vis image path: ", os.path.join(WKS_VIS_PATH, wks_vis_list[i]))
    # 将三张图拼接在一起可视化
    cv2_imshow(np.concatenate((gt_vis_img, cim_vis_img, wks_vis_img), axis=1))
    # break
    
```
